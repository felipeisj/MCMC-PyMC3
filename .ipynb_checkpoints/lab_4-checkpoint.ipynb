## Problema de la mochila


Dado un conjunto de $m$ elementos cada uno descrito por su masa $w_j$ y su valor $v_j$ y una mochila cuyo l铆mite de capacidad es $C$. Asumiendo que el volumen y la forma de los objetos no importan, encuentre el subconjunto de objetos de mayor valor que puede ser llevado en la mochila. Este es un problema de optimizaci贸n combinatorial NP completo

Podemos definir la variable indicadora

$$
x = (z_1, z_2, \ldots, z_m)
$$

donde cada $z_i \in \{0, 1\}$ es igual a 1 si el elemento $i$ est谩 en la mochila y 0 si no lo est谩

Se define entonces un espacio de posibilidades

$$
\Omega = \left \{x \in \{0, 1\}^m : \sum_{j=1}^m w_j z_j \leq C \right \}
$$

De donde queremos encontrar aquella que maximiza la utilidad

$$
U(x) = \sum_{j=1}^m v_j z_j
$$

## Soluci贸n con Monte Carlo

Para resolverlo con Monte Carlo podr铆amos

- Dado $x_t$
- Escoger $j \in [1, ..., m]$ al azar de manera uniforme
- Crear $y=(z_1, z_2, ..., 1-z_j,..., z_m)$, si es la mejor soluci贸n hasta ahora, guardarla
- Si $y$ es factible entonces $x_{t+1} = y$ de lo contrario $x_{t+1} = x_t$

Pero esto podr铆a tardar muchisimo para $m$ grande

## Actividad: Simulated Annealing

Resuelva este problema usando el algoritmo de Simulated Annealing, el cual es una versi贸n del algoritmo de Metropolis donde la distribuci贸n de inter茅s se formula a partir de la funci贸n de utilidad como

$$
p(x) = \frac{1}{Z} \exp \left(\frac{U(x)}{T} \right) 
$$

donde $T$ es la temperatura y $Z = \sum_{x\in \Omega}  \exp \left(\frac{U(x)}{T} \right)$ es la evidencia que no depende de $x$

```python


def lamb(r) : 
    return (min(1,r))

def r(y, x, T=100, X = X1) : 
    return (math.exp(mochila(x, "v", X)/T)/math.exp(mochila(y, "v", X)/T))
```

### Implemente la soluci贸n de Monte Carlo

```python
def mochila(x, valor, X = X1):
    suma = 0
    for i in range(len(x)):
        suma = suma + x[i]*X[valor][i]
    return suma
```

```python
x = [0] * 10
y=x.copy()
m=10
lista = []

for i in range(0,500000):
    j = randint(0, m-1)
    valor_old = mochila(x, "v")
    x[j] = 1 - x[j]
    if( mochila(x, "w") < X1["C"] and mochila(x, "v") > valor_old):
        lista.append(mochila(x, "v"))
        y=x.copy()
    else:
        x[j] = 1 - x[j]
plt.plot(lista)
plt.show()
print("El valor 贸ptimo es: ", mochila(y, "v"))
print("El peso 贸ptimo es: ", mochila(y, "w"))
```

### Implemente la soluci贸n de Simulated annealing modificando el criterio de aceptaci贸n ingenuo por de Metropolis

```python
def funcion_simulated_annealing(X_wwe, T=100, mix_time = 500000):
    
    us = scipy.stats.uniform.rvs(size=mix_time)
    x2 =[0] * len(X_wwe["v"])
    y2 = x2.copy()
    lista2 = []
    ar = 0.
    for n in range(1, mix_time):
        j = randint(0,len(X_wwe["v"])-1)
        valor_viejo = mochila(x2, "v", X_wwe)
        x2[j] = 1 - x2[j]
        valor = mochila(x2, "v", X_wwe)
        capacidad = mochila(x2, "w", X_wwe)

        if (mochila(x2, "w", X_wwe) < X_wwe["C"] and us[n] < lamb(r(y2,x2, T, X_wwe))):
            lista2.append(valor)
            y2 = x2.copy()
            ar += 1.
    print("El valor 贸ptimo es: ", mochila(y2, "v", X_wwe))
    print("El peso 贸ptimo es: ", mochila(y2, "w", X_wwe))
    return lista2, ar/mix_time
lista2, ar2 =funcion_simulated_annealing(X_wwe = X1)
plt.plot(lista2)
plt.show()

```

```python

def autocorr(thetas):
    thetas_norm = (thetas-np.mean(thetas))/np.std(thetas)
    rho = np.correlate(thetas_norm, 
                       thetas_norm, mode='full')
    return rho[len(rho) // 2:]/len(thetas)

def neff(rho):
    T = np.where(rho < 0.)[0][0]
    return len(rho)/(1 + 2*np.sum(rho[:T]))
```

```python
%%time
np.random.seed(12345)

fig, ax = plt.subplots(1, 2, figsize=(7, 3), tight_layout=True)
for sigma in [100, 350, 1000]:
    thetas, ar = funcion_simulated_annealing(X1, mix_time=2000, T=sigma)
    ax[0].plot(thetas)    
    ax[1].plot(autocorr(thetas), label=str(sigma))
    print(f"N煤mero efectivo de muestras para sigma {sigma}: {neff(autocorr(thetas)):0.4f}")
    print(f"Fracci贸n de aceptaci贸n: {ar:0.4f}")

ax[0].set_title('Traza')
ax[1].set_title('Autocorrelaci贸n');
ax[1].legend();
```
